## Q 34-01

Convierta los siguientes datos del diccionario en un dataframe usando **Pandas**.

(Denomina el dataframe `df`)

`d = { 'col1':[1,2], 'col2':[3,4], 'col3':[5,6], 'col4':[7,8] }`

```python
import pandas as pd
d = { 'col1':[1,2], 'col2':[3,4], 'col3':[5,6], 'col4':[7,8] }
df = pd.DataFrame(d)
df
```

## Q 37-01

Escriba un comando para buscar datos faltantes en el dataframe `df` creado en la pregunta **Q 34-01** e imprima el resultado.

(Sin embargo, los datos que faltan deben devolverse como verdaderos).

```python
df.isnull()
#Este metodo de panda devuelve un dataFrame del mismo tamano del original, donde cada celda contiene True si el valor NaN y falso de lo contrario
```

```python
df.notnull()
#Funciona de manera opuesta a isnull(), devolviendo True para los valores que no son NaN y False para los que lo son.
```
