```python
# Espacio disponible para desarrollar el código y agregar los comentarios necesarios.
import time
t_act = time.time() #Tiempo Actual en segundos desde el epoch
print("Tiempo actual en segundos", t_act)
t_seg = 1731011951.142441
t_list = time.gmtime(t_seg)  #este metodo devuelve la conversion de los segundos en una estructura de tiempo UTC
print("Tiempo en formato de lista", t_list)
t_legible = time.ctime(t_seg)
print("Tiempo en formato legible", t_legible)
time.strftime #formatea la estructura de tiempo en una cadena de caracteres segun el formato expresado
t_local = time.localtime(t_seg)
print("Tiempo local", t_local)
t_wait = time.sleep(5) #espera 5 segundos
print("Tiempo de espera 5 seg")
t_estructura = time.asctime(t_local)
print("Tiempo en formato de estructura", t_estructura)
```

```python
import datetime
fecha_actual = datetime.datetime.now() #devuelve el tiempo actual
print(fecha_actual)
fecha = datetime.datetime(2024, 11, 7) #permite construir un objeto desde una fecha ano, mes y dia
print(fecha)
fecha = datetime.datetime(2024, 11, 7, 10, 56, 24) #tambien se le pueden agregar horas minutos y segundos
print(fecha)
fecha_formato = datetime.datetime.strftime(fecha_actual, '%d/%m/%Y %H:%M:%S') #Este metodo permite cambiar el formato de la fecha
print(fecha_formato)
fecha_formato2 = datetime.datetime.strftime(fecha, '%B %m %Y %H:%M:%S') #con la b al inicio nos muestra el nommbre del mes
print(fecha_formato2)
fecha_Texto = 'Dec 06 2024 12:56:11'
fecha_actual_texto = datetime.datetime.strptime(fecha_Texto, '%b %d %Y %H:%M:%S') #Este metodo permite hacer un objeto desde una cadena de texto
print(fecha_actual_texto)
dia_time = int(datetime.datetime.strftime(fecha_actual, '%d')) #este nos devuelve el dia
print(dia_time)
hora_Actual = datetime.datetime.strftime(fecha_actual, '%H:%M:%S') #este nos devuelve la hora
print(hora_Actual)
utc_now = datetime.datetime.utcnow() #este nos devuelve la hora en UTC
print(utc_now)
combine_obj = datetime.datetime.combine(fecha.date(), fecha.time()) #este nos permite combinar una fecha y una hora
print(combine_obj)
time_Delta = datetime.timedelta()
print(time_Delta)
time_week = datetime.datetime.weekday(fecha) #este nos devuelve el dia de la semana
print(time_week)
time_replace = fecha.replace(year = 2023) #este nos permite cambiar el año de una fecha
print(time_replace)
time_iso = fecha.isoformat() #este nos devuelve la fecha en formato ISO
print(time_iso)
time_timestamp = fecha.timestamp() #este nos devuelve la fecha en formato timestamp
print(time_timestamp)
```

```python
lista_de_fechas = ['04/01/2024', '01/28/2024', '03/16/2024', '05/24/2024', '07/30/2025']
# Espacio disponible para desarrollar el código y agregar los comentarios necesarios.
fecha_datetime = [datetime.datetime.strptime(fecha, '%m/%d/%Y') for fecha in lista_de_fechas]
fecha_futura = datetime.datetime.strptime('07/30/2025', '%m/%d/%Y')
for fecha in fecha_datetime:
    dias_restantes = (fecha_futura - fecha).days
    semanas_restantes = dias_restantes // 7
    print(f"Fecha: {fecha.strftime('%Y-%m-%d')}, Semanas restantes: {semanas_restantes}, Días restantes: {dias_restantes}")
```

# Actividad: Manejo de la biblioteca Pandas.


## Instrucciones


1. Crear una serie de pandas, donde su indice este compuesto por fechas y en columna principal almace los valoras cargados en una lista.

2. Graficar los valores de la lista en función de la fecha.


```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.
import pandas as pd
import matplotlib.pyplot as plt
precios = [140, 130, 120, 115, 160, 110, 170, 150, 157, 135]
fechas = pd.date_range('2024-11-05', '2024-11-24', freq='2D')
print(fechas)
s = pd.Series(precios, index = [fechas])
print(s)
plt.plot(fechas, precios, marker='o', linestyle='-', color='b')
plt.title('Grafico de Precios') # Título del gráfico
plt.xlabel('Fecha') # Etiqueta para el eje x
plt.ylabel('Precio') # Etiqueta para el eje y
plt.xticks(rotation=45) # Rotar las etiquetas del eje x 45 grados
plt.grid()  # Mostrar la cuadrícula en el gráfico
plt.tight_layout() # Ajustar automáticamente el diseño del gráfico
plt.show() # Mostrar el gráfico
```
